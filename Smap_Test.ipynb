### 1. Read test files

```python
import pandas as pd
import matplotlib.pyplot as plt

smap_test = pd.read_csv('datasets/smap_test.csv')
smap_test_label = pd.read_csv('datasets/smap_test_label.csv')
```

### 2. Draw time series plots with anomaly regions:

```python
plt.plot(smap_test.index, smap_test.iloc[:, 1])  # Assuming the timestamps are not provided in the dataset
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Time Series Data')
```

### 3. Perform EDA and find out root cause

```python
plt.figure(figsize=(10, 6))
plt.hist(smap_test.iloc[:, 1], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Visualize the time series data
plt.figure(figsize=(12, 6))
plt.plot(smap_test.index, smap_test.iloc[:, 1], color='blue')
plt.title('Time Series Data')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

### 4. Find out the variables which are the root cause for the anomaly

```python
# Extract anomaly regions based on indices provided in the label file
anomaly_indices = smap_test_label.iloc[:, 0].tolist()
anomaly_values = smap_test.iloc[anomaly_indices, 1]

# Calculate summary statistics for anomaly regions
anomaly_stats = anomaly_values.describe()

# Calculate summary statistics for non-anomaly regions
non_anomaly_values = smap_test.drop(anomaly_indices, axis=0).iloc[:, 1]
non_anomaly_stats = non_anomaly_values.describe()

# Compare the statistics
print("Summary statistics for anomaly regions:")
print(anomaly_stats)
print("\nSummary statistics for non-anomaly regions:")
print(non_anomaly_stats)
```
